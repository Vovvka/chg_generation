```python
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import openpyxl


def get_excel_data ():
    file_path = f"path to file"

    workbook = openpyxl.load_workbook(file_path)

    # Select the sheet (active by default, modify if you need specific sheet)
    sheet = workbook.active

    row_num = input("Enter the row number to use: ")
        
    # Set column indices for the Specified columns
    collection_col = get_column_index(sheet, "Collection")
    description_col = get_column_index(sheet, "Description")
    start_time_col = get_column_index(sheet, "Start Time")
    time_zone_col = get_column_index(sheet, "Time Zone")
    duration_col = get_column_index(sheet, "Duration")
    Start_date_col = get_column_index(sheet, "Start Date")
    End_date_col = get_column_index(sheet, "End Date")


    # Get data from Specified columns in the chosen row
    Collection_name = sheet.cell(row=row_num, column=collection_col).value
    print(f"Start Description: {Collection_name}")
    start_description = sheet.cell(row=row_num, column=description_col).value
    print(f"Start Description: {start_description}")
    start_time = sheet.cell(row=row_num, column=start_time_col).value
    print(f"Start Time: {start_time}")
    start_timezone = sheet.cell(row=row_num, column=time_zone_col).value
    print(f"Start Timezone: {start_timezone}")
    duration = sheet.cell(row=row_num, column=duration_col).value
    duration_hours = int(duration.replace(" hours", ""))
    print(f"Duration (hours): {duration_hours}")
    start_date = sheet.cell(row=row_num, column=Start_date_col).value
    print(f"Duration (hours): {start_date}")
    end_date = sheet.cell(row=row_num, column=End_date_col).value
    print(f"Duration (hours): {end_date}")
    #add approval groups cell


def get_column_index(sheet, column_name):
    # Get the column index based on the column name
    for col_num in range(1, sheet.max_column + 1):
        if sheet.cell(row=1, column=col_num).value == column_name:
            return col_num
    raise ValueError(f"Column '{column_name}' not found in the sheet")


driver = webdriver.Chrome()

# Enter your login credentials and submit the form
def login_snow ():
    page = driver.find_element_by_id("user_name")
    page.send_keys("your_username")
    page = driver.find_element_by_id("user_password")
    page.send_keys("your_password")
    page.send_keys(Keys.RETURN)



def normal_cr ():

    change_type ="Normal"

    url = f"sysparm_collection_value={change_type}"

    driver.get(url)



def standard_cr ():

    change_type ="Standard"

    url = f"&sysparm_collection_value={change_type}"

    driver.get(url)


login_snow ()

```
