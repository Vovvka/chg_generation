```python
from datetime import datetime
import openpyxl

def get_data(output_file_path, row_num):
    try:
        workbook = openpyxl.load_workbook(output_file_path)
        output_sheet = workbook.active

        cr_number_col = get_column_index(output_sheet, "CR number")
        collection_col = get_column_index(output_sheet, "Collection")
        #Start_Date_col = get_column_index(output_sheet, "Start Date")
        #End_Date_col = get_column_index(output_sheet, "End Date")

        cr_number = output_sheet.cell(row=row_num, column=cr_number_col).value
        collection = output_sheet.cell(row=row_num, column=collection_col).value

        #parsed_start_date = output_sheet.cell(row=row_num, column=Start_Date_col).value.strftime("%m-%d-%Y %H:%M:%S")
        #parsed_end_date = output_sheet.cell(row=row_num, column=End_Date_col).value.strftime("%m-%d-%Y %H:%M:%S")

        fill_sfd(cr_number, collection)

    except Exception as e:
        print(f"An error occurred: {e}")

def get_column_index(sheet, column_name):
    for col_num in range(1, sheet.max_column + 1):
        if sheet.cell(row=1, column=col_num).value == column_name:
            return col_num
    raise ValueError(f"Column '{column_name}' not found in the sheet")

def fill_sfd(cr_number, collection):
    
    input_file_path = r'C:\Users\pelyu\OneDrive\Рабочий стол\chg\git\UMG - Syxsense Patch Calendar v357.xlsx'
    output_sheet_name = 'Scheduled for Deployment'

    # Read the existing data from the Excel file
    workbook = openpyxl.load_workbook(input_file_path)
    worksheet = workbook[output_sheet_name]

    # Get the column index for 'Group Name' and 'Change Request'
    group_name_col = None
    change_request_col = None
    for col_idx in range(1, worksheet.max_column + 1):
        if worksheet.cell(row=1, column=col_idx).value == 'Group Name':
            group_name_col = col_idx
        elif worksheet.cell(row=1, column=col_idx).value == 'Change Request':
            change_request_col = col_idx

    if group_name_col is None or change_request_col is None:
        print("Error: 'Group Name' or 'Change Request' column not found.")
        return

    # Update the 'Change Request' column for the relevant rows
    for row_idx in range(2, worksheet.max_row + 1):
        if worksheet.cell(row=row_idx, column=group_name_col).value == collection:
            worksheet.cell(row=row_idx, column=change_request_col).value = cr_number

    # Save the changes
    workbook.save(input_file_path)
     

if __name__ == "__main__":
    row_num = int(input("Enter the row: "))
    output_file_path = r'C:\Users\pelyu\OneDrive\Рабочий стол\chg\git\write_here.xlsx'
    get_data(output_file_path, row_num)
```
